```python
#catagorical plots
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset("tips")
tips.head()
```

```python
#(x=catagorical column,y=numerical colume)
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
#(extimator=default it is mean)
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#same as barplot but y is fixed with number of occurance
sns.countplot(x='sex',data=tips)
```

```python
#dot shows the outliar
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
#similar to boxplot but give more information like kernal density
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

```python
#stripplot made satterplot
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#it is the combination of strip and violin plot
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
#gernal method
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='point')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='box')
```

```python
#Grid
import seaborn as sns
import pandas as pd
%matplotlib inline
iris = pd.read_csv('C:\\Users\\hp\\Desktop\\iris_data.csv')
iris.head()
```

```python
iris['Species'].unique()
```

```python
iris.drop('Id',axis=1,inplace=True)
```

```python
iris.head()
```

```python
g=sns.PairGrid(iris)#it makes empty grids and the you map on them
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(iris)#it makes empty grids and the you map on them
g.map_diag(sns.distplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```
